# Angle of Detection for Different Cube Samples
In this notebook, we will calculate the angle of detection (`α`) for the 5 different samples with varying distances (`d`).
The detector has a height (`h`) of 17.9 cm above the sample. The relationship between these quantities is given by:


$$\text{hypotenuse} \times \cos(\alpha) = h$$


Let's proceed with the calculations.

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from math import acos, degrees

# Function to calculate angle alpha
def calculate_alpha(h, d):
    return degrees(acos(h / d))

# Function to plot the schematic diagram for a given d and alpha
def plot_schematic(h, d, alpha):
    # Coordinates for the triangle
    O = [0, 0]  # Origin
    A = [d * np.cos(np.radians(alpha)), 0]  # Sample location (base)
    B = [0, h]  # Detector location (height)
    
    # Setting up the plot
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plotting the corrected lines for the triangle
    ax.plot([O[0], B[0]], [O[1], B[1]], label='h (BO)', linewidth=2, color='g')  # h (height)
    ax.plot([O[0], A[0]], [O[1], A[1]], label='Base (OA)', linestyle='--', linewidth=2, color='b')
    ax.plot([A[0], B[0]], [A[1], B[1]], label=f'Hypotenuse (BA), d = {d} cm', linewidth=2, color='r')  # Hypotenuse
    
    # Marking and labeling the points
    for point, name in zip([O, A, B], ['O', 'A (Sample)', 'B (Detector)']):
        ax.text(point[0], point[1], f' {name}')
    
    # Adding labels for sides
    ax.text((B[0] + O[0])/2 - 0.5, (B[1] + O[1])/2 + 0.5, f'h = {h} cm', fontsize=12)
    
    # Drawing an arrow to indicate angle alpha
    arrow_start = [B[0], B[1]]
    arrow_end = [B[0] - 2.5, B[1] - 2.5]
    ax.annotate('', xy=arrow_end, xytext=arrow_start, arrowprops=dict(facecolor='black', arrowstyle='<|-|>', lw=1.5))
    
    # Adding annotation for angle alpha 
    angle_label_position = [B[0] - 3, B[1] - 3]
    ax.annotate(f'α = {alpha:.2f}°', xy=angle_label_position, fontsize=12)
    
    # Setting plot parameters
    ax.axis('equal')
    ax.grid(True)
    ax.legend()
    ax.set_title(f'Schematic Diagram for d = {d} cm')
    ax.set_xlabel('Base (cm)')
    ax.set_ylabel('Height (cm)')
    
    plt.show()
```

```python
# Given height of the detector
h = 17.9

# Given distances for each cube sample (cm)
cube_distances = {
    'Cube 1': 21.4,
    'Cube 2': 22.9,
    'Cube 3': 20.9,
    'Cube 4': 21.7,
    'Cube 5': 22.4
}
```

```python
# Calculating and plotting for each cube sample
for cube, d in cube_distances.items():
    alpha = calculate_alpha(h, d)
    print(f"The angle α for {cube} with d = {d} cm is {alpha:.2f}°")
    plot_schematic(h, d, alpha)
```
